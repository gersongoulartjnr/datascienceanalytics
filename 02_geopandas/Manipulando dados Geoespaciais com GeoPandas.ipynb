## Biblioteca GeoPandas para Manipulação e Visualização de Dados Geoespeciais

Este notebook tem como objetivo apresentar a biblioteca GeoPandas para manipulação e visualização de dados geoespaciais.

GeoPandas é a principal biblioteca de análise de dados geoespaciais em python, devido a similaridade com o pacote pandas. O GeoPandas estende os tipos de dados usados pelos pandas para permitir operações espaciais em tipos geométricos. As operações geométricas são realizadas pelo pacote shapely. Para acessar arquivos, pode-se utilizar a bibliteca fiona e matplotlib para plotagem.

Abaixo serão listados algumas das principais funcionalidades do pacote.

Para mais detalhes acessar: https://geopandas.org/en/stable/

#### 1. Estrutura dos dados

A estrutura de dados central em GeoPandas é o **GeoDataFrame**, uma subclasse de pandas.DataFrame, que pode armazenar colunas geométricas e realizar operações espaciais. Um **GeoDataFrame** deve possui colunas de dados e necessariamente uma coluna de geometria, através da estrutura **GeoSeries**. Cada GeoSeries pode conter qualquer tipo de geometria - pontos, linhas, polígonos que podem estar em um único array - e possui um atributo CRS, que armazena informações sobre o sistema de referência de coordenadas.

#### 2. Tipos de Geometrias do GeoPandas

Os objetos de geometria, que são do tipo GeoSeries, tem três classes básicas de objetos geométricos (que utilizam a biblioteca `shapely`):
- Pontos;
- Linhas;
- Polígonos

Esses objetos podem ser combinados através de arrays (como é o exemplo do multiponto, multilinhas e multipolígonos).

##### 2.7. Atributos e Métodos
Os principais atributos dos objetos geométricos são:
- `area`: área da forma
- `bounds`: tupla de coordenadas máximas e mínimas em cada eixo para cada forma
- `total_bounds`: tupla de coordenadas máximas e mínimas em cada eixo para toda a GeoSeries;
- `geom_type`: tipo de geometria;
- `is_valid`: testa se as coordenadas formam uma forma geométrica razoável.

Os principais métodos e testes dos objetos geométricos são:
- `distance()`: distância entre dois objetos;
- `centroid()`: centróide de dois objetos;
- `plot()`: plot da geometria;
- `geom_almost_equals()`: retorna se os objetos são quase idênticos (solução quando temos precisão de pontos flutuantes que podem deixar as formas ligeiramente diferentes);
- `contains()`: retorna se um objeto está contido em outro.
- `intersects()`: retorna se há intersecção entre os objetos.

#### 3. Lendo um GeoDataFrame

Um GeoDataFrame pode ser obtido através de um arquivo de dados e geometria (GeoPackage, GeoJSON, Shapefile) ou convertido de um dataframe do Pandas.


Os arquivos de dados e geometria que possuem informações dos polígonos que delimitam as regiões são conhecidos como `Shapefiles`. São arquivos compactados (.zip) que possuem informações de geometria, tabela de informações, dados de indexação e coordenadas.

Outro formato comum de se encontrar Shapefiles é através de um arquivo KML (Keyhole Markup Language), uma linguagem desenvolvida para ser usada com o Google Earth, baseada em XML.

##### 3.1. Arquivo de dados de Polígonos no formato ShapeFile

Neste exemplo temos um arquivo compactado (.zip) que possui informações do shapefile dos bairros da cidade de Nova York.

```python
import geopandas as gpd
path_to_data = gpd.datasets.get_path("nybb")
gdf = gpd.read_file(path_to_data)
gdf.head(3)
```

```python
gdf.plot()
```

##### 3.3. Convertendo um DataFrame do Pandas que contém latitude e longitudade em um GeoDataFrame 

Neste exemplo temos um dataset de remessas que possui latitude e longitude. Dessa forma, é necessário que o dataframe esteja no formato do pandas.

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, Polygon, mapping
df = pd.read_csv('data/dataframe.csv')
df.head(3)
```

Após, é necessário através da biblioteca shapely transformar as coordenadas em pontos. Assim como mencionado anteriormente, o dataframe deve possuir apenas uma coluna de geometria.

```python
geometry = [Point(xy) for xy in zip(df['lng'], df['lat'])]
geo_df = gpd.GeoDataFrame(df, geometry = geometry)
geo_df.head(3)
```

#### 4. Merging de dados
Existem duas maneiras de combinar conjuntos de dados em geopandas: junções de atributos e junções espaciais.

##### 4.1. Junções de Atributos
As junções de atributos são realizadas usando o método `merge()`. É recomendável usar o método merge chamado do conjunto de dados espaciais. 

Caso for usada em um dataframe comum do pandas, o merge funcionará se o GeoDataFrame estiver no argumento esquerdo. Se um DataFrame estiver no argumento esquerdo e um GeoDataFrame estiver na posição direita, o resultado não será mais um GeoDataFrame.

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
country_shapes = world[['geometry', 'iso_a3']]
country_shapes.head(3)
```

```python
cities = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
country_names = world[['name', 'iso_a3']]
country_names.head(3)
```

```python
country_shapes = country_shapes.merge(country_names, on='iso_a3')
country_shapes.head(3)
```

##### 4.1. Junções Espaciais
Em uma União Espacial, as observações de dois GeoSeries ou GeoDataFrame são combinadas com base em seu relacionamento espacial entre si e podem ser de 2 tipos: Predicate Joins e Nearest Joins.

##### 4.1.1. Predicate Joins
No Predicate Join o argumento do predicate especifica como geopandas decide se deve ou não unir os atributos de um objeto a outro, com base em seu relacionamento geométrico.

Possuem dois argumentos: `predicate` e `how`.

O argumento `predicate` especifica como geopandas decide se deve ou não unir os atributos de um objeto a outro, com base em seu relacionamento geométrico e podem ser do tipo:
- **intersects**;
- **contains**;
- **within**;
- **touches**;
- **crosses**;
- **overlaps**

O argumento `how` especifica o tipo de junção que ocorrerá e qual geometria será retida no GeoDataFrame resultante. 
Segue o padrão de um join relacional, podendo ser: **inner**, **right** ou **left**.

```python
countries = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
cities = gpd.read_file(gpd.datasets.get_path("naturalearth_cities"))
cities_w_country_data = gpd.sjoin(left_df = countries, right_df = cities, how="inner")
cities_w_country_data.head(3)
```

##### 4.1.2. Nearest Joins

O nearest joins como seu próprio nome diz, realiza joins através de objetos que estejam próximos. Também possui o argumento how, assim como os Predicate Joins.

Nesse tipo de join é importante informar dois argumentos: **max_distance** (o raio de distância entre os objetos) e **distance_col** (acrescenta uma coluna com o resultado do cálculo das distâncias).

##### 5.1. Adicionando contexto (background) aos mapas do GeoPandas

Para usar background dos mapas do geopandas é necessário da biblioteca contextily. A biblioteca oferece backgrounds de diversas fontes com focos diferentes (ruas, relevo, etc).

É importante que o mapa e o contexto possuam o mesmo sistema de coordenadas.

Para saber mais, acesse a documentação: https://contextily.readthedocs.io/en/latest/

```python
import geopandas as gpd
import contextily as cx

df = gpd.read_file(gpd.datasets.get_path('nybb'))
ax = df.plot(alpha=0.5, edgecolor='k')
```

```python
#Alterando o sistema de coordenadas de referência
df_wm = df.to_crs(epsg=3857)

ax = df_wm.plot(figsize=(6, 6), alpha=0.5, edgecolor='k')

#Adicionando o contexto
cx.add_basemap(ax, source=cx.providers.OpenStreetMap.Mapnik)

ax.set_axis_off()
```
